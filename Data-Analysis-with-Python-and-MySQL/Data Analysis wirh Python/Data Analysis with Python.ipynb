```python
df['Year'].value_counts().plot(kind="pie",autopct='%1.1f%%')
plt.title("Yearly Distribution")
plt.axis('equal')
plt.savefig("Yearly Distribution.png")
plt.show()
```

```python
df.groupby('JobTitle')['BasePay'].mean().sort_values().tail(10).plot(kind="bar", color=plt.cm.viridis(np.linspace(0, 1, 10)))
plt.title("Top 10 Job Titles by Average BasePay")
plt.xlabel("Job Title")
plt.ylabel("Average Base Pay")
plt.xticks(rotation=90)
plt.savefig("Top 10 Job Titles by Average BasePay.png")
plt.show()
```

```python
df.groupby("JobTitle")['OvertimePay'].mean().sort_values().tail(10).plot(kind="bar", color=plt.cm.viridis(np.linspace(0, 1, 10)))
plt.title("Top 10 Job Titles by Average overtimepay")
plt.xlabel("Job Title")
plt.ylabel("Average overtimepay")
plt.xticks(rotation=90)
plt.savefig("Top 10 Job Titles by Average overtimepay")
plt.show()
```

```python
df.groupby("JobTitle")['TotalPayBenefits'].mean().sort_values().tail(10).plot(kind="bar", color=plt.cm.viridis(np.linspace(0, 1, 10)))
plt.title("Top 10 Job Titles by Average TotalPayBenefits")
plt.xlabel("Job Title")
plt.ylabel("TotalPayBenefits")
plt.xticks(rotation=90)
plt.savefig("Top 10 Job Titles by Average TotalPayBenefits.png")
plt.show()
```

```python
df.groupby('Year')['TotalPayBenefits'].mean().plot(kind="bar")
plt.title("Average Total Pay and Benefits by Year")
plt.xlabel("Year")
plt.ylabel("Total Pay and Benefits")
plt.savefig("Average Total Pay and Benefits by Year.png")
plt.show()
```

```python
cols = ['BasePay', 'OvertimePay', 'OtherPay', 'TotalPay', 'TotalPayBenefits']
avg_values_by_year = df.groupby('Year')[cols].mean().reset_index()
avg_values_melted = avg_values_by_year.melt(id_vars='Year', var_name='Metric', value_name='AverageValue')
sns.set_style("whitegrid")
plt.figure(figsize=(12, 8))
sns.barplot(x='Year', y='AverageValue', hue='Metric', data=avg_values_melted, palette='husl')
plt.title('Average Values by Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Value', fontsize=14)
plt.legend(title='Metric', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig("Average Values by Year.png")
plt.show()
```
